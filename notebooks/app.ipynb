```python
import pandas as pd
import ipywidgets
import folium
from folium.plugins import MarkerCluster
from ipyleaflet import Map

data = pd.read_csv("DATAX/HOSP_VAC_FALLECIDOS_NUEVO.csv")


data['fecha_ingreso_hosp'] = pd.to_datetime(data['fecha_ingreso_hosp'])

data = data.sort_values(by='fecha_ingreso_hosp')

#print(data['fecha_ingreso_hosp'])

diccionario = {}

for row in data.iterrows():
    array = []
    res = not bool(diccionario)
    contador = 0
    list_of_key = list(diccionario.keys())

    if(row[1].fecha_ingreso_hosp != None):
        fecha_ingreso_hosp = row[1].fecha_ingreso_hosp
        #print(str("Fecha antes: "+ str(fecha_ingreso_hosp)))
        fecha_ingreso_hosp = pd.to_datetime(fecha_ingreso_hosp, dayfirst=True, errors='ignore')
        fecha_ingreso_hosp = fecha_ingreso_hosp.strftime("%Y/%m")
        #print(str("Fecha despues: "+ fecha_ingreso_hosp))
    if(row[1].Latitude != None):
        lat = row[1].Latitude
        new_lat = lat.replace('.','')
        new_lat = new_lat[:3]  + '.' + new_lat[3:]
    if(row[1].Longitude != None):
        long = row[1].Longitude
        new_long = long.replace('.','')
        new_long = new_long[:3]  + '.' + new_long[3:]
    
    new_lat_long = str(new_lat + "," + new_long)
    array.append(new_lat_long) 
    if (res):
        diccionario[fecha_ingreso_hosp] = array
    else:
        if(fecha_ingreso_hosp in list_of_key):
            array = list(diccionario[fecha_ingreso_hosp])
            array.append(new_lat_long)
            diccionario[fecha_ingreso_hosp] = array
        else:
            diccionario[fecha_ingreso_hosp] = array
        array = []

df = pd.DataFrame(list(diccionario.items()), columns=['Date', 'Lat_Long'])
selection_slider = ipywidgets.Dropdown(
    options=list(df['Date']),
    #value='2',
    description='Date:',
    disabled=False,
)

#e = df.loc[df['Date'] == '08/2020']
#for(index,row) in e.iterrows():
    #for values in row.loc['Lat_Long']:
        #print(values)


def plot_data(date):
    e = df.loc[df['Date'] == date]
    #print(e)
    for(index,row) in e.iterrows():
        map_osm = folium.Map(location=[-10.462564, -75.465432], zoom_start=7)
        mc = MarkerCluster()
        for lat_long in row.loc['Lat_Long']:
            new_lat_long = str(lat_long).split(",")
            latitude = new_lat_long[0]
            longitude = new_lat_long[1] 

            #print("NEWLATLONG: " + str(new_lat_long[1]))
            #print("lat: " + str(latitude))
            #print("long " + str(longitude))
            
            mc.add_child(folium.Marker(location=[latitude, longitude], popup=row.loc['Date']))
            #marker = Marker(location=[row.loc['lat'], row.loc['long']])
            #covid_map.add_layer(MarkerCluster(markers=marker))
            map_osm.add_child(mc)       
        display(map_osm)  
          
ipywidgets.interact(plot_data, date=selection_slider)
```
